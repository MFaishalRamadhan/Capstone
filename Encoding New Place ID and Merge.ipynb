```python
import pandas as pd
```

```python
# Load place dataset
from google.colab import drive
drive.mount("/content/gdrive")
place_dataset = pd.read_excel(r'/content/gdrive/My Drive/Colab Notebooks/Dataset/Dataset Tempat.xlsx')
place_dataset.head()
```

```python
# Load review user dataset
from google.colab import drive
drive.mount("/content/gdrive")
rating_dataset = pd.read_excel(r'/content/gdrive/My Drive/Colab Notebooks/Dataset/Dataset Rating Users.xlsx')
rating_dataset.head()
```

```python
place_id_mapping = {}
new_place_id_counter = 1
```

```python
for index, row in place_dataset.iterrows():
    existing_place_id = row['Place_ID']
    place_id_mapping[existing_place_id] = new_place_id_counter
    new_place_id_counter += 1
```

```python
# Replace place IDs in place_data using new place IDs
place_dataset['Place_ID'] = place_dataset['Place_ID'].map(place_id_mapping)
```

```python
place_dataset
```

```python
# Replace place IDs in review_data using new place IDs
rating_dataset['Place_ID'] = rating_dataset['Place_ID'].map(place_id_mapping)
```

```python
rating_dataset
```

```python
users_dataset = rating_dataset.merge(place_dataset, on='Place_ID')
```

```python
users_dataset.to_excel('Users_Dataset.xlsx', index=False)
```
